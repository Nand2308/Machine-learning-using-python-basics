# Matplotlib
<pre><b>Data Visualization Technique</pre>

```python
import matplotlib.pyplot as plt
```

```python
#Line Plot
x = [2,6,9,5,4,2,9,7]
y = [6,3,2,7,8,4,8,1]

plt.plot(x,y,'*-.',color='yellow')
plt.plot(x,y,'g')
plt.title('Line Plot',color='Blue')
plt.xlabel('X-Axis',color='c')
plt.ylabel('Y-Axis',color='c')
plt.grid(color='pink')
plt.show()
```

```python
#Bar Plot
x = [1,3,5,7,9]
y = [2,4,6,8,10]

plt.bar(x,y,color=['y','pink'],label='Example 1')
plt.grid(color='pink')
plt.show()
x1=[4,7,2,9,8]
y1=[9,2,6,4,1]
plt.bar(x1,y1,color=['y','c'],label='Example 2')
plt.grid(color='pink')
plt.show()
plt.title('Bar Graph')
plt.legend()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(color='pink')
```

# HISTOGRAM

```python
#Histogram
ages=[2,18,15,8,10,12,20,25,35]
bins=[0,10,20,30,40,50]
plt.hist(ages,bins,histtype='bar',rwidth=0.8,color='violet',edgecolor='y',linewidth=3)
plt.title('Hsitogram')
plt.show()
```

```python
#Scatter plot
x = [3,5,4,7,2,10,7,8]
y = [2,6,4,9,6,9,3,10]
plt.scatter(x,y,s=500,color='indigo',marker='*',edgecolor='y',)
plt.title('Scatter Plot')
plt.show()
```

```python
#Scatter plot
x = [3,5,4,7,2,10,7,8]
y = [2,6,4,9,6,9,3,10]
plt.bar(x,y,color='brown',width=0.05)

plt.scatter(x,y,s=500,color='indigo',marker='*',edgecolor='y',linewidths=12)
plt.scatter(x,y,color='c')
plt.title('Scatter Plot')
plt.show()
```

```python
#pie chart
slices = [10,8,3,5,4]
activities = ['Phone','Sleeping','Studying','Playing','Eating']
cols = ['b','gold','r','m','g']
plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0,0.3,0,0),autopct='%1.3f%%')
plt.title('Pie chart')
plt.show()
```
