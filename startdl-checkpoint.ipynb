### Activation function
<li>it decides whether a neuron should be activated or not by calculating the weighted sum and further adding the bias to it.</li>
<li><b> types of Activation function</li>
<li> Linear (op: continuous) </li>
<li> Sigmoid op: categorical </li> 
<li> TanH (Hyperbolic Tangent) (op: categorical)</li>
<li> ReLU (Rectified linear unit)(for both continuous and categorical) </li> 

<b>* Linear <b>
<li> Formula: y = mx + c

<b>* Sigmoid <b>
<li> Formula: y = 1/(1+e^-x)
<li>range: (0,1)</li>

<b>* TanH<b>
<li> Formula: y = (1-e^-2x)/(1+e^-2x)
<li>range: (-1,+1)</li>

<b>*ReLU<b>
<li>range: (0,infinity)
<li>it converts the -ve values to zero and keep the +ve value as it is</li>
