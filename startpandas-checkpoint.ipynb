```python
#Creating a DataFrame
#By using dictionary we can create a DataFrame
d={'Name':['Ram','Siya','Nandini','Jyoo','Vishnu'],'Roll':[111,222,333,444,555],'Course':['AIML','Python','AIML','AIML','AIML']}
d
```

```python
df = pd.DataFrame(d,index=[1,2,3,4,5])
df
```

```python
df=pd.DataFrame(d,index=range(1,6))
df
```

```python
import numpy as np
d = pd.DataFrame({'Days':[1,2,3,4,5],'Places':['Puri',np.NaN,'Munnar','Siripuram','Kerala'],'Visitors':[2000,3000,6000,1000,5000]})
d
```

```python
#to check the misssing values
d.isnull()
```

```python
d.isnull().sum()
```
